```python
import pandas as pd
from collections import Counter
import nltk
import re
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
```

```python
def preprocess(post):
    preprocessedWords = []
    lemmatizer = WordNetLemmatizer()
    
    post = post.lower()
    post = re.sub(r'[^\w\s]', '', post)
    words = post.split()
    for word in words:
        lemmatized = lemmatizer.lemmatize(word)
        preprocessedWords.append(lemmatized)
    
    return preprocessedWords
```

```python
df = pd.read_json('../data/postdatalines.json', lines=True, orient='records')

keywords = ['depression','i']
wordCountsSent = []

for index, row in df.iterrows():
    post = row['text'][0]
    preppedWords = preprocess(post)
    wordCountsPost = dict(Counter(preppedWords))
    keywordCounts = {}
    for word in keywords:
        try:
            keywordCounts[word] = [wordCountsPost[word],index]
        except: 
            continue
    wordCountsSent.append(keywordCounts)
    
  
    
    


```
