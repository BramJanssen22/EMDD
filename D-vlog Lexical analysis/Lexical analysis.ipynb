```python
import pandas as pd
from collections import Counter
import nltk
import re
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
df = pd.read_json('../data/postdatalines.json', lines=True, orient='records')
keywordsjson = pd.read_json('../data/depression_synonyms.json', orient='records')
keywords =  list(keywordsjson['depression'])
print(keywords)
```

```python
def preprocess(post):
    preprocessedWords = []
    lemmatizer = WordNetLemmatizer()
    
    post = post.lower()
    post = re.sub(r'[^\w\s]', '', post)
    words = post.split()
    for word in words:
        lemmatized = lemmatizer.lemmatize(word)
        preprocessedWords.append(lemmatized)
    
    return preprocessedWords
```

```python
wordCountsPosts = []

for index, row in df.iterrows():
    post = row['text'][0]
    preppedWords = preprocess(post)
    wordCountsPost = dict(Counter(preppedWords))
    keywordCounts = {}
    keywordCounts['id'] = index
    for word in keywords:
        try:
            keywordCounts[word] = wordCountsPost[word]
        except: 
            continue
    wordCountsPosts.append(keywordCounts)

print(wordCountsPosts)
```

```python
wordCountsTotal = {}

for word in keywords:
    wordCountsTotal[word] = 0

for wordCountsPost in wordCountsPosts:
    for word in keywords:
        try:
            wordCountsTotal[word] = wordCountsTotal[word] + wordCountsPost[word]
        except:
            continue
    wordCountsTotal['id'] = 'total'
```

```python
def totaldepressionwords(depressionDict):
    total = 0
    for word in keywords:
        try:
            total += depressionDict[word]
        except:
            continue
    return depressionDict['id'], total
```
